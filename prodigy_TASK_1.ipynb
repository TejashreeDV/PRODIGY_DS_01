```python


# Sample data
np.random.seed(0)
data = {'Age': np.random.randint(18, 70, size=100)}
df = pd.DataFrame(data)

# Plot the histogram
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=10, color='blue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {'Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female', 'Female', 'Male', 'Female']}
df = pd.DataFrame(data)

# Count the number of occurrences of each category
gender_counts = df['Gender'].value_counts()

# Plot the bar chart
plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color=['skyblue', 'grey'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```
